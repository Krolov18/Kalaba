# Exemple d'une grammaire prenant en compte les ambiguité morphologique ET syntaxique
Avec un vocabulaire comme le/la/les/chat/chats/cuisine/cuisines/dort/dorment/dans, on peut vite exploser la combinatoire.

```python
def generate_grammar_productions(g, depth):
    memo = set()
    for x in generate(g, depth=depth):
        if (s := " ".join(x)) in memo:
            continue
        memo.add(s)
        trees = [*parser.parse(x)]
        if trees:
            print(
                pd.DataFrame(
                    {s: [" ".join(tree.label()["KA", "Traduction"])
                         for tree in trees]}
                ).to_markdown()
            )
            print()
```

```python
def pprint(id_print: str, s: str, trees: Iterator[Tree]) -> Any:
    """ factory pour afficher de différentes façons les arbres d'un parsing """
    concrete_product = f"print_{id_print}"
    package = __name__
    return factory_function(concrete_product=concrete_product,
                            package=package,
                            trees=trees,
                            s=s)


def print_possible_trees(trees: Iterator[Tree], s: str) -> None:
    """
    pour chaque arbre, on dessine l'arbre associé (style HPSG)
    Attention, il faut quitter l'arbre, pour que le suivant apparaisse.
    Il faut donc quitter le dernier arbre à afficher pour quitter cette fonction (ou tuer le process qui exécute cette fonction)
    """
    print(
        pd.DataFrame(
            {s: [tree.draw() for tree in trees]}
        ).to_markdown()
    )


def print_traduction(trees: Iterator[Tree], s: str) -> None:
    """ pour chaque arbre, on affiche les traductions associées """
    print(
        pd.DataFrame(
            {s: [" ".join(tree.label()["KA", "Traduction"])
                 for tree in trees]}
        ).to_markdown()
    )


def print_syntagmes(trees: Iterator[Tree], s: str) -> None:
    """ pour chaque arbre, on affiche les syntagmes sous formes parenthésée """
    print(
        pd.DataFrame(
            {s: [str(tree.label()["KA", "Syntagmes"])
                 for tree in trees]}
        ).to_markdown()
    )
```

## Les deux exemples suivants sont ambigus morphologiquement
### La phrase suivante est ambiguë syntaxiquement

```python
s = "les chats dans la cuisine dorment"
trees = [*parser.parse(s.split(" "))]
pprint(id_print="syntagmes", s=s, trees=trees)
print()
pprint(id_print="traduction", s=s,trees=trees)
```

### La phrase suivante n'est pas ambiguë syntaxiquement

```python
s = "les chats dorment dans la cuisine"
trees = [*parser.parse(s.strip().split(" "))]
pprint(id_print="syntagmes", s=s ,trees=trees)
print()
pprint(id_print="traduction", s=s ,trees=trees)
```
